# Symbology

This Notebook showcases the advanced symbology expressions provided in Python

```python
# Simple fill and stroke color

from jupytergis import GISDocument, constant

doc = GISDocument(latitude=47.0000, longitude=2, zoom=5)

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geojson_layer(
    path="data/france_regions.geojson",
    symbology=[
        constant("green").encoding("fill"),
        constant("white").encoding("stroke"),
        constant(1).encoding("stroke-width"),
    ],
)

doc
```

```python
# GeoJSON symbology with a color ramp

from jupytergis import ClassificationMode, GISDocument, field

doc = GISDocument()

doc.add_raster_layer(
    url="https://server.arcgisonline.com/ArcGIS/rest/services/Canvas/World_Light_Gray_Base/MapServer/tile/{z}/{y}/{x}",
)

doc.add_geojson_layer(
    path="data/eq.geojson",
    symbology=[
        field("mag").encoding("radius"),
        field("felt")
        .color_ramp(
            "viridis",
            domain=[1, 9000],
            mode=ClassificationMode.LOGARITHMIC,
            n_shades=10,
        )
        .encoding("fill"),
    ],
)

doc
```

```python
# Conditional rendering
# Use `when(...).constant(...)` and `when(...).field(...)` for readable guarded chains

from jupytergis import GISDocument, between, field, when

doc = GISDocument()

doc.add_raster_layer(
    url="https://server.arcgisonline.com/ArcGIS/rest/services/Canvas/World_Light_Gray_Base/MapServer/tile/{z}/{y}/{x}",
)

doc.add_geojson_layer(
    path="data/eq.geojson",
    symbology=[
        field("mag").encoding("radius"),
        when(field("mag") >= 8).constant("red").encoding("fill"),
        # Equivalent to when(field("mag") > 3, field("mag") < 8).constant("orange").encoding("fill"),
        when(between("mag", 3, 8)).constant("orange").encoding("fill"),
        when(field("mag") <= 3).constant("green").encoding("fill"),
    ],
)

doc
```

```python
# Conditional rendering can be used for filtering

from jupytergis import GISDocument, field, when

doc = GISDocument()

doc.add_raster_layer(
    url="https://server.arcgisonline.com/ArcGIS/rest/services/Canvas/World_Light_Gray_Base/MapServer/tile/{z}/{y}/{x}",
)

doc.add_geojson_layer(
    path="data/eq.geojson",
    symbology=[
        when(field("mag") >= 5).field("mag").identity().encoding("radius"),
        when(field("mag") >= 5).constant("red").encoding("fill"),
    ],
)

doc
```

```python
# Heatmap support

from jupytergis import GISDocument, field, heatmap

doc = GISDocument(latitude=57.676696, longitude=11.864487, zoom=12)

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")
doc.add_geojson_layer(
    name="Ships Trajectories Heatmap",
    path="https://openlayers.org/en/latest/examples/data/geojson/ship-trajectories.json",
    symbology=[
        heatmap(
            radius=1,
            blur=5,
            mappings=[field("$density").color_ramp("hot").encoding("pixel-color")],
        ),
    ],
)
doc
```

```python
# GeoTIFF symbology applied on a pixel basis

from jupytergis import GISDocument, field

doc = GISDocument(latitude=63.787048, longitude=-45.249785, zoom=6)

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geotiff_layer(
    url="https://s2downloads.eox.at/demo/EOxCloudless/2020/rgbnir/s2cloudless2020-16bits_sinlge-file_z0-4.tif",
    min=2000,
    max=25000,
    symbology=[
        field("$band-1")
        .scalar(domain=(0, 0.5), output_range=(0, 1))
        .encoding("pixel-alpha"),
        field("$band-1").color_ramp("winter", n_shades=9).encoding("pixel-rgb"),
    ],
)

doc
```

```python
# GeoTIFF symbology single band

from jupytergis import GISDocument, field

doc = GISDocument(latitude=48.103308, longitude=15.662690, zoom=5)

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geotiff_layer(
    url="https://eoresults.esa.int/d/FCM-AGB-100m/2023/01/01/FCM-AGB-100m-2023/FCM_Europe_demo_2023_AGB.tif",
    min=0,
    max=240,
    symbology=[
        field("$band-1")
        .identity()
        .encoding("pixel-red", "pixel-green", "pixel-blue", "pixel-alpha"),
    ],
)

doc
```

```python
# GeoTIFF symbology single band (custom channels)

from jupytergis import GISDocument, constant, field

doc = GISDocument(latitude=48.103308, longitude=15.662690, zoom=5)

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geotiff_layer(
    url="https://eoresults.esa.int/d/FCM-AGB-100m/2023/01/01/FCM-AGB-100m-2023/FCM_Europe_demo_2023_AGB.tif",
    min=0,
    max=240,
    symbology=[
        constant(0).encoding("pixel-red"),
        field("$band-1")
        .scalar(domain=(0, 1), output_range=(0, 255))
        .encoding("pixel-green"),
        constant(0).encoding("pixel-blue"),
        field("$band-1").identity().encoding("pixel-alpha"),
    ],
)

doc
```
