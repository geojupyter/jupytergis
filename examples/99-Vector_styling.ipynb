```python
# Vega expressions, Python expressions to OpenLayers expressions.

from jupytergis import GISDocument

doc = GISDocument()


doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")


doc.add_geojson_layer(
    path="data/france_regions.geojson",
    name="France Python Expr",
    python_expr=(
        "'purple' if datum.code > 80 else"
        "'blue' if datum.code > 60 else"
        "'orange' if datum.code > 40 else"
        "'cyan' if datum.code > 20 else"
        "'yellow'"
    ),
)

doc.add_geojson_layer(
    path="data/eq.geojson",
    vega_expr="""datum.mag > 6 ? 'red' :
                 datum.mag > 5 ? 'orange' :
                 datum.mag > 4 ? 'yellow' :
                 datum.mag > 3 ? 'cyan' :
                 datum.mag > 2 ? 'pink' :
                 '#00FF00'""",
)


doc
```
