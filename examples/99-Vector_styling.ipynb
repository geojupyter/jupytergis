```python
from jupytergis import GISDocument

doc = GISDocument()

doc.add_raster_layer(url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geojson_layer(
    path="data/france_regions.geojson",
    vega_expr="""
    datum.code > 80 ? 'red' :
    datum.code > 60 ? 'blue' :
    datum.code > 40 ? 'green' :
    datum.code > 20 ? 'cyan' :
    'pink'
    """
)

doc.add_geojson_layer(
    path="data/eq.geojson",
    vega_expr="""datum.mag > 6 ? 'red' : 
                 datum.mag > 5 ? 'orange' : 
                 datum.mag > 4 ? 'yellow' : 
                 datum.mag > 3 ? 'cyan' :
                 datum.mag > 2 ? 'pink' : 
                 '#00FF00'"""
)


doc
```

```python
doc.add_vectortile_layer(
    url="https://basemaps.arcgis.com/arcgis/rest/services/World_Basemap_v2/VectorTileServer/tile/{z}/{y}/{x}.pbf",
    color_expr={"fill-color": "#00FF00", "circle-fill-color": "#FF0000"},
)
```
